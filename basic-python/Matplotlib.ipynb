```python
x = np.arange(1, 11)
y = x ** 2

plt.plot(x, y, color='blue', marker='o', linestyle='--', label='x squared')
plt.title("Line Plot Example")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color='purple')
plt.title("Scatter Plot Example")
plt.xlabel("Random X")
plt.ylabel("Random Y")
plt.show()
```

```python
subjects = ['Math', 'Science', 'English', 'History']
marks = [88, 92, 80, 75]

plt.bar(subjects, marks, color='orange')
plt.title("Bar Graph Example")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```

```python
plt.barh(subjects, marks, color='teal')
plt.title("Horizontal Bar Graph")
plt.show()
```

```python
activities = ['Study', 'Sleep', 'Exercise', 'Leisure']
hours = [6, 8, 2, 8]

plt.pie(hours, labels=activities, autopct='%1.1f%%', startangle=90)
plt.title("Daily Routine")
plt.show()
```

```python
data = np.random.normal(50, 10, 200)  # mean=50, std=10

plt.hist(data, bins=10, color='skyblue', edgecolor='black')
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
data = [np.random.normal(50, 10, 100), np.random.normal(60, 5, 100)]

plt.boxplot(data)
plt.title("Box Plot Example")
plt.xlabel("Group")
plt.ylabel("Values")
plt.show()
```

```python
x = np.arange(1, 6)
y1 = [2, 3, 4, 5, 6]
y2 = [1, 2, 3, 4, 5]

plt.stackplot(x, y1, y2, labels=['Series1', 'Series2'], colors=['blue', 'lightblue'])
plt.title("Area Plot Example")
plt.legend()
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, axs = plt.subplots(2)
axs[0].plot(x, y1, 'r')
axs[0].set_title("Sine Wave")

axs[1].plot(x, y2, 'b')
axs[1].set_title("Cosine Wave")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create fake temperature data for 7 days × 5 cities
temperature = np.array([
    [30, 32, 34, 33, 31],
    [31, 33, 35, 34, 32],
    [32, 34, 36, 35, 33],
    [33, 35, 37, 36, 34],
    [34, 36, 38, 37, 35],
    [33, 35, 36, 35, 34],
    [32, 34, 35, 33, 32]
])

days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
cities = ['Delhi', 'Mumbai', 'Chennai', 'Kolkata', 'Bengaluru']

plt.imshow(temperature, cmap='hot', interpolation='nearest')
plt.title("Weekly Temperature Heatmap (°C)")
plt.xlabel("Cities")
plt.ylabel("Days")
plt.xticks(np.arange(len(cities)), cities)
plt.yticks(np.arange(len(days)), days)
plt.colorbar(label='Temperature (°C)')
plt.show()
```

```python
plt.imshow(temperature, cmap='hot')
plt.title("Temperature Heatmap with Values")
plt.xlabel("Cities")
plt.ylabel("Days")
plt.xticks(np.arange(len(cities)), cities)
plt.yticks(np.arange(len(days)), days)

for i in range(len(days)):
    for j in range(len(cities)):
        plt.text(j, i, temperature[i, j], ha='center', va='center', color='white')

plt.colorbar(label='Temperature (°C)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
temperature = np.random.randint(15, 45, (12, 4))

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
           'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
cities = ['Delhi', 'Mumbai', 'Chennai', 'Kolkata']

plt.imshow(temperature, cmap='hot', interpolation='nearest')
plt.title("Average Monthly Temperature (°C)")
plt.xlabel("Cities")
plt.ylabel("Months")
plt.xticks(np.arange(len(cities)), cities)
plt.yticks(np.arange(len(months)), months)
plt.colorbar(label="Temperature (°C)")
plt.show()
```
