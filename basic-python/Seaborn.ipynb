```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
print(df.head())
```

```python
df.species.value_counts()
```

```python
sns.kdeplot(data=df, x='sepal_length', hue='species', fill=True)
```

```python
sns.scatterplot(data=df, x="sepal_length", y="sepal_width", hue="species", style="species")
plt.title("Sepal Length vs Width")
plt.show()
```

```python
sns.pairplot(df, hue="species")
plt.suptitle("Iris Pairwise Feature Relationships", y=1.02)
plt.show()
```

```python
sns.violinplot(data=df, x="species", y="petal_width", inner="quartile")
plt.title("Petal Width Distribution by Species")
plt.show()
```

```python
corr = df.drop('species', axis=1).corr()

sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
sns.swarmplot(data=df, x="species", y="sepal_length", palette="Set2")
plt.title("Sepal Length per Species")
plt.show()
```
